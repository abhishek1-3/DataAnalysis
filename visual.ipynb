```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

#Line Plot
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Line Plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

#Line Plot
plt.plot(x,y,color='red',linestyle='-.',marker='X',linewidth=2,markersize=15)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title("Line Plot")
plt.grid(True)
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,14,5,77,32]

plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title('Plot 1')

plt.subplot(2,2,2)
plt.plot(y1,x,color='red')
plt.title('Plot 1')

plt.subplot(2,2,3)
plt.plot(x,y2,color='blue')
plt.title('Plot 1')

plt.subplot(2,2,4)
plt.plot(y2,x,color='yellow')
plt.title('Plot 1')
plt.show()
```

```python
#Bar Plot
categories=['A','B','C','D','E']
values=[5,7,8,9,6]

plt.bar(categories,values,color='purple')
plt.title('Bar plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
# Histogram   Distribution of data
data=[1,2,3,4,3,5,3,5,5,3,5,3,5,2,1,4,5,3]

plt.hist(data,bins=5,color='orange',edgecolor='black')
```

```python
#Scatter plot
x=[1,2,3,4,5]
y=[2,3,4,6,7]

plt.scatter(x,y,marker='*')
```

```python
# pie chart
labels=['a','b','c','d']
sizes=[15,20,40,30] 
colors=['gold','green','blue','lightcoral']
explode=(0.2,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```

```python
#Sales Data Visulaization
import pandas as pd
df=pd.read_csv('data/Details.csv')
```

```python
df.head(3)
```

```python
#total profit by product
total_profit_by_product=df.groupby('Sub-Category')['Profit'].sum()
total_profit_by_product
```

```python
total_profit_by_product.plot(kind='bar',color='teal')
```

```python
#According to payment mode total profit
Profit_wrt_paymentMode=df.groupby('PaymentMode')['Profit'].sum().reset_index()
```

```python
Profit_wrt_paymentMode
```

```python
plt.plot(Profit_wrt_paymentMode['PaymentMode'],Profit_wrt_paymentMode['Profit'],color='red')
```

#### SEABORN 

```python
import seaborn as sns
```

```python
#Basic Plotting with seaborn
tips=sns.load_dataset('tips')
tips
```

```python
#Create Scatterplot
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('Total bill Vs Tip')
plt.show()
```

```python
#line plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('Total bill vs size')
plt.show()
```

```python
# Bar 
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar plot of total bill per day')
plt.show()
```

```python
#BOX PLOT
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
##Violin plot
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
#histogram
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
#KDE plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
#PairPlot
sns.pairplot(tips)
```

```python
#heatmap
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
