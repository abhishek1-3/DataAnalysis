```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
```

```python
df=pd.read_csv('data/business.csv')
df.head()
```

```python
df.describe()
```

```python
#feature engg process
df['Date']=df['date'].str.split('-').str[0]
df['Month']=df['date'].str.split('-').str[1]
df['Year']=df['date'].str.split('-').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.drop('date',axis=1, inplace=True)
```

```python
df['Price']=df['price'].str.replace(',',"")
df['Price']=df['Price'].astype(int)
df.drop('price',axis=1,inplace=True)
```

```python
df['Arrival_hour']=df['arr_time'].str.split(':').str[0]
df['Arrival_min']=df['arr_time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
df.drop('arr_time', axis=1, inplace=True)
```

```python
df['Departure_hour']=df['dep_time'].str.split(':').str[0]
df['Departure_min']=df['dep_time'].str.split(':').str[1] 
```

```python
df['Departure_hour']=df['Departure_hour'].astype(int)
df['Departure_min']=df['Departure_min'].astype(int)
df.drop('dep_time', axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df['stop'].unique()
```

```python
df['Stop']=df['stop'].str.split('-').str[0]
df['Stop'].unique()
```

```python
df['Stop']=df['Stop'].map({'non':0,'1':1,'2+':2})
df.drop('stop', axis=1,inplace=True)
df.head(3)
```

```python
df['hour_taken']=df['time_taken'].str.split(' ').str[0].str.split('h').str[0]
df['min_taken']=df['time_taken'].str.split(' ').str[1].str.split('m').str[0]
df.drop('time_taken',axis=1,inplace=True)
```

```python
df['hour_taken']=df['hour_taken'].astype(int)
df['min_taken']=df['min_taken'].astype(int)
```

```python
df['Duration(min)']=(df['hour_taken']*60)+df['min_taken']
```

```python
df.drop(['hour_taken','min_taken'], axis=1, inplace=True)
```

```python
encoder=OneHotEncoder()
encoder.fit_transform(df[['airline','from','to','ch_code']] ).toarray()
```

```python
data=pd.DataFrame(encoder.fit_transform(df[['airline','from','to','ch_code']] ).toarray(),columns=encoder.get_feature_names_out())
```

```python
f_data=pd.concat([df,data],axis=1)
f_data.head(3)
f_data.drop(['ch_code','from','to','airline'], axis=1,inplace=True)
```
