```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('data/business.csv')
df.head()
```

```python
df.describe()
```

```python
#feature engg process
df['Date']=df['date'].str.split('-').str[0]
df['Month']=df['date'].str.split('-').str[1]
df['Year']=df['date'].str.split('-').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.drop('date',axis=1, inplace=True)
```

```python
df['Price']=df['price'].str.replace(',',"")
df['Price']=df['Price'].astype(int)
df.drop('price',axis=1,inplace=True)
```

```python
df['Arrival_hour']=df['arr_time'].str.split(':').str[0]
df['Arrival_min']=df['arr_time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
df['Arrival_hour']=df['Arrival_hour'].astype(int)
df['Arrival_min']=df['Arrival_min'].astype(int)
df.drop('arr_time', axis=1, inplace=True)
```

```python
df['Departure_hour']=df['dep_time'].str.split(':').str[0]
df['Departure_min']=df['dep_time'].str.split(':').str[1] 
```

```python
df['Departure_hour']=df['Departure_hour'].astype(int)
df['Departure_min']=df['Departure_min'].astype(int)
df.drop('dep_time', axis=1, inplace=True)
```

```python
df.head(2)
```

```python
df['stop'].unique()
```

```python
df['Stop']=df['stop'].str.split('-').str[0]
df['Stop'].unique()
```

```python
df['Stop']=df['Stop'].map({'non':0,'1':1,'2+':2})
df.drop('stop', axis=1,inplace=True)
df.head(3)
```
